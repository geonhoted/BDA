```python
import pandas as pd

df = pd.read_csv(r'C:\Users\장건호\Desktop\BDA\taiwanese+bankruptcy+prediction\data.csv')

print(df.head())
print(df.info())
print(df.describe())
```

```python
# 결측치 확인
print(df.isnull().sum())

# 결측치 시각화 (heatmap)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
missing_rows = df.isnull().any(axis=1).sum()
print(f'결측치가 포함된 행 수: {missing_rows}')
```

```python
df.columns
```

```python
# 각 열의 기본 통계 정보
print(df.describe())

# 수치형 데이터 분포 (히스토그램)
df.hist(bins=20, figsize=(20, 15), grid=False)
plt.show()

# 타겟 변수인 'Bankrupt?'의 분포 확인
plt.figure(figsize=(10, 6))
sns.countplot(x='Bankrupt?', data=df)
plt.title('Bankruptcy Count')
plt.show()
```

```python
# 상관계수 행렬
corr_matrix = df.corr()

# 상관계수 히트맵 시각화
plt.figure(figsize=(20, 18))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()

# 타겟 변수와의 상관관계 확인
target_corr = corr_matrix['Bankrupt?'].sort_values(ascending=False)
print('Bankrupt?와의 상관관계:')
print(target_corr)
```

```python
# 타겟 변수와의 상관관계가 높은 상위 10개 특징 확인
important_features = target_corr.head(11).index
print('타겟 변수와의 상관관계가 높은 상위 10개 특징:')
print(important_features)

# 중요한 특징들에 대한 시각화 (박스플롯 예시)
for feature in important_features[1:]:
    plt.figure(figsize=(5, 3))
    sns.boxplot(x='Bankrupt?', y=feature, data=df)
    plt.title(f'Boxplot of {feature} by Bankruptcy')
    plt.show()
```
